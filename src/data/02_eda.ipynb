```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
df = pd.read_csv("../../data/raw/california_housing.csv")
df.head()
```

```python
df.info()
df.describe()
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Feature Correlation Heatmap")
plt.tight_layout()
plt.savefig("../../docs/eda/feature_correlation_heatmap.png", dpi=300)
plt.show()
```

```python
sns.histplot(df["MedHouseVal"], bins=50, kde=True)
plt.title("Distribution of Median House Value")
plt.xlabel("Median House Value ($100k)")
plt.ylabel("Count")
plt.savefig("../../docs/eda/distribution_median_house_value.png", dpi=300)
plt.show()
```

```python
df.hist(figsize=(14, 10), bins=50)
plt.suptitle("Feature Distributions", fontsize=16)
plt.tight_layout()
plt.savefig("../../docs/eda/feature_distributions.png", dpi=300)
plt.show()
```
